6) Realice un histograma con la población de los Municipios para cada una de las provincias.

````python
def histogramas_por_provincia(df):
    provincias = df.Provincia.unique()
    
    for provincia in provincias:
        datos_provincia = df[df.Provincia == provincia]
        
        # Histograma de la provincia
        plt.figure(figsize=(10, 6))
        plt.hist(datos_provincia.Poblacion, alpha=0.7)
        plt.title(f'Distribucion de Poblacion en {provincia}')
        plt.xlabel('Poblacion')
        plt.ylabel('Cantidad de Municipios')
        
        # dibujar histograma
        plt.show()
````

````python
histogramas_por_provincia(df)
````

8) Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

````python
# utilizamos el municipio de Madrid como base para nuestro análisis
datos_madrid = df[df.Municipio == 'Madrid']
datos_madrid
````

````python
madrid_x = datos_madrid.X.values[0]
madrid_y = datos_madrid.Y.values[0]
print("Madrid se encuentra en las coordenadas", madrid_x, "," , madrid_y)
````

````python
# obtenemos un dataframe excluyendo a Madrid
df_distancias = df.copy()
df_distancias = df_distancias[ df.Municipio != 'Madrid' ]

# sacampos la distancia con base a la formula matematica
df_distancias['Distancia_a_Madrid'] = np.sqrt(
    (df.X - madrid_x)** 2 + (df.Y - madrid_y)** 2
)
````

````python
# determinar el municipio más cercano
municipio_mas_cercano = df_distancias.loc[df_distancias.Distancia_a_Madrid.idxmin()]['Municipio']
print(f"El municipio más cercano es {municipio_mas_cercano}.")
````

````python
# determinar el municipio más lejano
municipio_mas_lejano = df_distancias.loc[df_distancias.Distancia_a_Madrid.idxmax()]['Municipio']
print(f"El municipio más lejano es {municipio_mas_lejano}.")
````
